```python
import numpy as np
import pandas as pd
df = pd.read_csv("googleplaystore.csv")
```

```python
df.head()
df.columns.str.strip()
```

```python
df.info()
```

```python
df
```

```python
df['Installs'] = df['Installs'].str.replace('+', '').str.replace(',', '', regex=True)
df['Size'] = df['Size'].str.replace('M', '', regex=True)
```

```python
df
```

```python
df['Rating']=pd.to_numeric(df['Rating'],errors='coerce')
df['Reviews']=pd.to_numeric(df['Reviews'],errors='coerce').astype('Int64')
df['Size']=pd.to_numeric(df['Size'],errors='coerce')
df['Installs']=pd.to_numeric(df['Installs'],errors='coerce').astype('Int64')
df['Price']=pd.to_numeric(df['Price'],errors='coerce')
```

```python
df.info()
```

```python
df['App'] = df['App'].str.lower()
df['Category'] = df['Category'].str.lower()
df['Type'] = df['Type'].str.lower()
df['Content Rating'] = df['Content Rating'].str.lower()
df['Genres'] = df['Genres'].str.lower()
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Rating'] = df['Rating'].fillna(0)
df['Reviews'] = df['Reviews'].fillna(0)
df['Size'] = df['Size'].fillna(0)
df['Installs'] = df['Installs'].fillna(0)
df['Type'] = df['Type'].fillna("undefined")
df['Price'] = df['Price'].fillna(0)
df['Content Rating'] = df['Content Rating'].fillna("undefined")
df['Current Ver'] = df['Current Ver'].fillna("undefined")
df['Android Ver'] = df['Android Ver'].fillna("undefined")
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(keep=False, inplace=True)
```

```python
df.duplicated().any()
```

```python
df[df["App"].str.contains("astrology" , case = False)]["App"].count()
```
